```python
# liste silme 
my_list = [1,2,3,4,5,6,7,8,9,20]
print(f"orjinal liste : {my_list}")
my_list.clear()
print(f"silinmis l,20iste: {my_list}\n\n")

# liste sonuna eleman ekler
my_list.append(10)
print(f"liste sonuna eleman ekler: {my_list}\n\n")

# listedeki ilgili elemanin sayisini dondurur
my_list = [10,20,30,10,40,10,50,60,70,80,10,20,10]
recurring_cntr = my_list.count(10)
print(f"listede 10 dan {recurring_cntr} adet mevcut\n\n")

#listeleri birlestirir
print(f"my list :{my_list}\t patch list : {patch_list}")
patch_list = [1,0,0,1,0,0,1,0,0,1]
my_list.extend(patch_list)
print(f"extended list: {my_list}\n\n")

#listedeki ilgili indekse ilgili sayiyi yazar
my_list.insert(0,100)
print(f"0. indekse 100 yaz: {my_list}\n\n")

# indeksteki ilgili degeri kaldirir
kaldirilan=my_list.pop(0)
print(f"{kaldirilan} listede  100leri kaldir : {my_list}\n\n")

# listeyi kopyalar Shallow copy
clone_list = my_list.copy()
print(f"my_list    : {my_list}")
print(f"clone_list : {clone_list}")
# orijinalde degisiklik yapildiginde shallow clone degismez
my_list.append(12032000)
print(f"my_list after append    : {my_list}")
print(f"clone_list after append : {clone_list}\n\n")

# listeyi ters cevirir
print(f"my_list                  : {my_list}")
my_list.reverse()
print(f"my_list after reverse    : {my_list}\n\n")


# listeyi kucukten buyuge siralar
my_list=[10,9,23,56,78,91,76,25,21,7,100,58]
print(f"my_list                  : {my_list}")
my_list.sort()
print(f"my_list after sortinf    : {my_list}\n\n")

#listedeki ilgili degeri cikarir
print(f"my_list                  : {my_list}")
my_list.remove(9)
print(f"my_list                  : {my_list}")

#listedeki ilgili degeri cikarir
print(f"my_list                  : {my_list}")
print(f"10'un indeksi            : {my_list.index(10)}")
```

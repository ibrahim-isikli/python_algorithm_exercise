### ARNOLDS CAT MAP


    formul : 
    [x']  = M [x]  mod (N)
    [y']      [y] 


    M -> Determinanti 1 olan herhangi bir matris ama genelde bu yontemde [1 1]
                                                                         [1 2]
                                                                     

```python
import numpy as np

def arnolds_cat_map(koordinat, N):
    M = np.array([[1,1],[1,2]])
    sonuc_matrisi = (np.dot(M,koordinat)) % N
    print("Sonuc Matrisi:")
    print(sonuc_matrisi)
    # RETURN sonuc matrisi


# user param
N = 5  #boyutu
default_koordinat_matrisi = np.array([[1,2],[2,3]])
arnolds_cat_map(default_koordinat_matrisi, N)
```

### BERNOULLI SHIFT MAP

x_n+1  = (2*x_n) mod(1)

```python
def bernoulli_shift_map(x_0,iterations):
    # init start value
    sequence = [x_0]
    for _ in range(iterations):
        x_next = (2*sequence[-1]) % 1
        print(f"[{_+1}] x = {x_next}\n")
        sequence.append(x_next)
    return x_next

# user params
x_0 = 0.35
iterations = 3
sequence = bernoulli_shift_map(x_0,iterations)
print(f"\n\ninit val :{x_0}\t max_iter:{iterations}\t result:{sequence}")
```

### CHEBYSHEV MAP

    x_n+1 = cos(a * cos^-1(x_n))
    -2 <= a <= 2

    

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
k = 2.0  # constant
x0 = 0.5  # initial condition
iterations = 100  # number of iterations

# Initialize the list to store the values of x
x_values = [x0]

# Generate the trajectory using the Chebyshev map
for i in range(iterations):
    x_next = np.cos(k * np.arccos(x_values[-1]))  # Chebyshev map equation
    x_values.append(x_next)

# Plot the trajectory
plt.figure(figsize=(10, 6))
plt.plot(x_values, label='Chebyshev Map Trajectory')
plt.title('Chebyshev Map Trajectory')
plt.xlabel('Iteration')
plt.ylabel('x')
plt.grid(True)
plt.legend()
plt.show()
```

### Circle Map
x_n+1 = x_n + omega - (k/2*pi) * sin(2*pi*x_n)

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
omega = 1.0  # frequency
K = 0.5  # coupling constant
x0 = 0.1  # initial condition
iterations = 1000  # number of iterations

# Initialize the list to store the values of x
x_values = [x0]

# Generate the trajectory using the Circle map
for i in range(iterations):
    x_next = (x_values[-1] + omega + K * np.sin(2 * np.pi * x_values[-1])) % 1
    x_values.append(x_next)

# Plot the trajectory
plt.figure(figsize=(10, 6))
plt.plot(x_values, label='Circle Map Trajectory')
plt.title('Circle Map Trajectory')
plt.xlabel('Iteration')
plt.ylabel('x')
plt.grid(True)
plt.legend()
plt.show()
```

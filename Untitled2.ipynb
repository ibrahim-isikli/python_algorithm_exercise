```python
import pygame
import random

# Pygame başlat
pygame.init()

# Ekran boyutları
WIDTH, HEIGHT = 400, 600
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Flappy Bird Klonu")

# Renkler
WHITE = (255, 255, 255)
BLUE = (0, 150, 255)
GREEN = (0, 255, 0)

# Kuş ayarları
bird_x, bird_y = 50, HEIGHT // 2
bird_radius = 15
bird_velocity = 0
gravitational_force = 0.5
jump_strength = -8

# Boru ayarları
pipe_width = 50
pipe_gap = 150
pipe_x = WIDTH
pipe_height = random.randint(100, 400)
pipe_speed = 3

# Skor
score = 0
font = pygame.font.Font(None, 36)

# Oyun döngüsü
running = True
clock = pygame.time.Clock()
while running:
    screen.fill(WHITE)
    
    # Olayları kontrol et
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        if event.type == pygame.KEYDOWN and event.key == pygame.K_SPACE:
            bird_velocity = jump_strength
    
    # Yerçekimi uygula
    bird_velocity += gravitational_force
    bird_y += bird_velocity
    
    # Boru hareketi
    pipe_x -= pipe_speed
    if pipe_x + pipe_width < 0:
        pipe_x = WIDTH
        pipe_height = random.randint(100, 400)
        score += 1
    
    # Çarpışma kontrolü
    if bird_y - bird_radius <= 0 or bird_y + bird_radius >= HEIGHT:
        print("Oyun Bitti! Skor:", score)
        running = False
    if pipe_x < bird_x + bird_radius < pipe_x + pipe_width:
        if bird_y - bird_radius < pipe_height or bird_y + bird_radius > pipe_height + pipe_gap:
            print("Oyun Bitti! Skor:", score)
            running = False
    
    # Çizimler
    pygame.draw.circle(screen, BLUE, (bird_x, int(bird_y)), bird_radius)
    pygame.draw.rect(screen, GREEN, (pipe_x, 0, pipe_width, pipe_height))
    pygame.draw.rect(screen, GREEN, (pipe_x, pipe_height + pipe_gap, pipe_width, HEIGHT))
    
    score_text = font.render(f"Skor: {score}", True, (0, 0, 0))
    screen.blit(score_text, (10, 10))
    
    pygame.display.update()
    clock.tick(30)

pygame.quit()
```
